# Python Data Variables
***


Variables in Python language are programing segments, like recipients that store a piece of information. It could be a number, a string of text, or even a complex data structure. Think of variables as labeled boxes that hold different types of data. In Python, declaring a variable is easy . You don’t need to specify the type explicitly like in some other programming languages. Python is smart enough to figure it out on its own, thanks to its dynamic typing system as the following Built-in Data Types. Examples:

###### # Declare a variable called *'name'* and assign it a string value
name = "Joe Smith"

###### # declare a variable called *'age'* and assign it an integer value
age = 25

###### # Declare a variable called *'is_student'* and assign it a boolean value
is_student = True

* As for what _types_ of variables we should use, for our purposes floats for numbers shall be used in order to have increased accuracy (and the values come as floats from the dataset given), and strings for strings of characters (such as sex, species, island etc). Note that for the flipper length variable, we do not need to use float values as all values have nothing past the decimal point, however, for the purposes of consistency in our values, we will keep them as such.

## Load Data
***

##### Load the palmer penguins data set from a URL.

```python
# Load the penguins data set
df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv")
```

##### The data in now loaded below so we can inspect it.

```python
# drops all NaN instances.
# source: https://stackoverflow.com/questions/44548721/remove-row-with-null-value-from-pandas-data-frame
df = df.dropna(how='any',axis=0) 

# Let's have a look.
df
```

# Two Variables Plots
***


```python
# Get only the bill lengths.

blen = df['bill_length_mm']

# Show.
print(blen)

# Type
print(type(blen))
```

```python
# Get only the numpy array.

blen = blen.to_numpy()

# Show.
blen
```

```python
# Bill depths
bdepth = df['bill_depth_mm'].to_numpy()

# Show.
bdepth
```

```python
# Simple plot
plt.plot(blen,bdepth, '*')

# Axis Labels.
plt.xlabel('bill length (mm)')
plt.ylabel('bill depth (mm)')

# Title.
plt.title('Penguins Data Set')

# X Limits.
plt.xlim(30, 60)

# Y Limits.
plt.ylim(12.5, 23)

```

```python
# create aa new figure and set of axes.
fig, ax = plt.subplots()

# Simple plot.
ax.plot(blen, bdepth, '*')

# Axis Labels.
ax.set_xlabel('bill length (mm)')
ax.set_ylabel('bill depth (mm)')

# Title
ax.set_title('Penguins Data Set')

# X Limits
ax.set_xlim(30, 60)

# Y Limits
ax.set_ylim(12.5, 23)


```

#### Add a best fit line.
***

$ y = mx + c = p_1 x^1 + p_0 = p_1 x + p_0$

```python
# Fit a straight line between x and y.
m, c = np.polyfit(blen, bdepth, 1)

# Show m and c.
m, c
```

```python
# Create a new figure and set of axes.
fig, ax = plt.subplots()

# Simple plot.
ax.plot(blen, bdepth, '*')

# Simple plot.
ax.plot(blen, m * blen + c, 'r-')

# Axis labels.
ax.set_xlabel('bill Length (mm)')
ax.set_ylabel('bill depth (mm)')

# Title.
ax.set_title('Penguins Data Set')

# X limits.
ax.set_xlim(30, 60)

# Y limits.
ax.set_ylim(12.5, 23)
```

```python
# x values for best fit line.
bf_x = np.linspace(30.0, 60.0, 1000)

# y values for best fit line.
bf_y = m * bf_x + c
```

```python
# Create a new figure and set of axes.
fig, ax = plt.subplots()

# Simple plot.
ax.plot(blen, bdepth, '*')

# Simple plot.
ax.plot(bf_x, bf_y, 'r-')

# Axis labels.
ax.set_xlabel('bill Length (mm)')
ax.set_ylabel('bill Width (mm)')

# Title.
ax.set_title('Penguins Data Set')

# X limits.
ax.set_xlim(30, 60)

# Y limits.
ax.set_ylim(10, 24)
```

```python
# Measure the correlation.
np.corrcoef(blen, bdepth)
```

# Creating  Bar Charts

***


```python
# Grouping the the data by 'species' and counting the occurrences

grouped_data = df['species'].value_counts()

# creating a bar chart
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', color='blue')

# Adding labels and titles
plt.title('Number of Penguins by Species')
plt.xlabel('Species')
plt.ylabel('Count')

# Rotating x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Groupin the data by 'species' and 'sex' and counting the occurrencies.
grouped_data = df.groupby(['species' ,'sex']).size().unstack()

# Creating a bar chart.
plt.figure(figsize=(10, 6))
grouped_data.plot(kind='bar', color=['skyblue', 'salmon'])

# Adding labels and title.
plt.title('Count of Penguins by Species and Sex')
plt.xlabel('Species')
plt.ylabel('Count')

# Rotating x-axis for better readability.
plt.xticks(rotation=45)

# Displaying the plot
plt.tight_layout()
plt.show()
```

# Creating  Histograms
***


```python
# Selecting the numerical variable for the histogram.
variable_for_histogram = 'body_mass_g'

# Creating the histogram.
plt.figure(figsize=(10, 6))
plt.hist(df[variable_for_histogram], bins=20, color='lightblue', edgecolor='black')

# Adding labels and title
plt.title('Histogram of Body Mass')
plt.xlabel('Body Mass (g)')
plt.ylabel('Frequency')

# Displaying the plot
plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Load the penguins dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv"
penguins_data = pd.read_csv(url)

# Drop rows with missing values in flipper_length_mm
penguins_data.dropna(subset=['flipper_length_mm'], inplace=True)

# Create a histogram of flipper length
plt.figure(figsize=(10, 6))
plt.hist(penguins_data['flipper_length_mm'], bins=20, color='skyblue', edgecolor='black')

# Add labels and title
plt.title('Histogram of Flipper Length')
plt.xlabel('Flipper Length (mm)')
plt.ylabel('Frequency')

# Add grid for better readability
plt.grid(True)


# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Calculate Pearson's correlation coefficient in Python

correlation_coefficient = penguins_data['body_mass_g'].corr(penguins_data['flipper_length_mm'])

print("Pearsons's correlation coefficient:", correlation_coefficient)
```
